# Fourth Neural Network

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

Despues de realizar la importacion de las 3 librerias:
Tensorflow, libreria creada por Google para trabajar el Machine Learning .

Numpy manejo de arreglos numericos.

Matpltlib para realizar la graficacion de curvas de entrenamiento de la red neuronal.




```python
in_meters=np.array([1, 2, 5, 10, 15, 20, 30, 40, 100, 400, 1000], dtype=float)
out_kilometres = np.array([0.001, 0.002, 0.005, 0.010, 0.015, 0.02, 0.03, 0.04, 0.1, 0.4, 1], dtype=float)
```

# Topologia de Neural Network

Keras es un Framework que nos permite realizar redes neurolanes de manera simple. Dense es el tipo de conexion entre neuronas.

```python
label1=tf.keras.layers.Dense(units=1, input_shape=[1])
```

Modelo de la red Secuencial

```python
model=tf.keras.Sequential([label1])
```

## Proceso de entrenamiento de la red 

```python
model.compile(
    optimizer = tf.optimizers.Adam(0.4),
    loss = 'mean_squared_error'
)
```

```python
print("Se esta realizando el entrenamiento de la red, Wait a minute, please")
```

```python
training = model.fit(in_meters, out_kilometres, epochs=880, verbose=False)
```

```python
model.save('DataNeural.h5')
model.save_weights('weights.h5')
```

```python
i=input("Ingrese un valor en metros, para ser convertido en Kilometres: ")
i=float(i)
prediction=model.predict([i])
print("El valor en kilometros es: ", str(prediction))
```

```python
plt.title("Grafica de aprendizaje Machine Learning")
plt.xlabel("Ciclos de entrenamiento")
plt.ylabel("Proceso de errores")
plt.plot(training.history["loss"])
plt.show()

print("Terminamos de realizar el entrenamiento de la red. ")
```
